```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Сравнение библиотек ###

```
crc32 = pd.DataFrame({
    'Word Count' : [ 27543, 387599 ],
    'CRC++, s' : [ 0.0030839, 0.040069 ],
    'Fast_CRC32 fast, s' : [ 0.0033489, 0.0487425 ],
    'Fast_CRC32 16bytes_prefetch, s' : [ 0.0045488, 0.0553833 ],
    'Fast_CRC32 crc32_16bytes, s' : [ 0.0051831, 0.0711648 ]
})

crc32.index = [ 'Small', 'Large' ]

crc32
```

```
tbl1 = pd.DataFrame({
    'Duration' : [ 9.85e-05, 0.0002123, 0.0005641, 0.0006964, 0.0010142, 0.0013462, 0.0046499, 0.0190667, 0.0265917 ],
    'Word count' : [ 387, 1937, 3875, 7751, 11627, 19379, 38759, 193799, 387599 ],
    'Dimension' : [ 246, 718, 1098, 1712, 2120, 2898, 4363, 10276, 14528 ]
})
```

### Размерность от количества слов ###

```
plt.figure(figsize=(15, 10))

plt.title('График зависимости размерности в-ра на выходе от кол-ва слов на входе', fontsize=18)
plt.xlabel('Размерность в-ра', fontsize=16)
plt.ylabel('Кол-во слов', fontsize=16)

plt.plot(tbl1['Word count'][:-1], tbl1['Dimension'][:-1], 'o-')
```

```
plt.figure(figsize=(15, 10))

plt.title('График зависимости размерности в-ра на выходе от кол-ва слов на входе', fontsize=18)
plt.xlabel('log(Размерность в-ра)', fontsize=16)
plt.ylabel('log(Кол-во слов)', fontsize=16)

plt.plot(tbl1['Word count'][:-1], tbl1['Dimension'][:-1], 'o-')
plt.xscale('log')
plt.yscale('log')
```

Получилась степенная зав-ть $y = x^k$.

### Время работы от количества слов ###

```
plt.figure(figsize=(15, 10))

plt.title('График зависимости времени работы от кол-ва слов на входе', fontsize=18)
plt.xlabel('Размерность в-ра', fontsize=16)
plt.ylabel('Скорость работы, сек', fontsize=16)

plt.plot(tbl1['Word count'][:-3], tbl1['Duration'][:-3], 'o-')
```
